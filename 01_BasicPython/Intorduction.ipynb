# Python Programming

## Introduction

This course is intended to give introduction to the Python programming language. It focusses on giving a bief overview of basic language concept to enable the adience to use it to solve computational tasks and problems. It is not considered a comprehensive coverage of the language.

### Course Requirements

This course requires a working Jupyter environment to actively execute the code blocks and solve small programming task and exercises.

### Course Format

The course is presented as a set of Jupyter notebooks that includes some exercised and taks for self learning.

### Learning Objectives

After this course you should have learned and be able to:

- run and edit a Jupyter notebook
- be familiar with Python syntax
- write and execute mathematical expressions
- handle conditional code execution
- apply control flow with loops
- know collection data types and how to use them
- understand how to use functions to structure your program

### Outline

Part 1: Variables and primitive data types

Part 2: Operators

Part 3: Control flow

Part 4: Collection data types

Part 5: Functions, modules and error handling

Part 6: Advanced patterns (classes, coroutines, iterators)
