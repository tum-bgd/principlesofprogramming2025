```python
def is_prime(x):
    for i in range(2, x): # better go to the sqrt(x) which you dont compute, but you square the other.
        if x % i == 0:
            return False

    return True

primes = [x for x in range(20) if is_prime(x)]
print(primes)
```

# Task: Generate a fixed number of prime numbers
- while loop: go ahead step by step and check if something is prime
- essentially do a loop, but over ranges (1...100, 100 ...200,...) "batched"


```python
# while version
primes = set()
i = 2
while len(primes) != 100:
    if is_prime(i):
        primes.add(i)
    i = i + 1
print(primes)
```

```python
# batched version:
for i in range(5):
    a = i * 10
    b = (i+1)*10
    print(a,b)
    print([x for x in range(a,b) if is_prime(x)])
```

# Good Python Loops
THere is a way to have a loop run for a long time, but not at once. 



```python
def get_primes(input=None):
    # this shall generate one prime at a time using as little time as possible
    if input is None:
        input= range(2,10000000)
    for i in input:
        if (is_prime(i)):
            yield i 

for i,p in enumerate(get_primes()):
    if i > 100:
        break
    print(p, end=" ")
```

```python
class Counter:
    def __init__(self, length):
        self.length = length
        self.i = 0
    def __iter__(self):
        return self
    def __len__(self):
        return self.length
    def __next__(self):
        time.sleep(1)
        self.i = self.i+1
        if self.i > self.length:
            raise StopIteration
        return self.i
        
c = Counter(10)
print(len(c))
for v in tqdm(c):
    pass
    #print(v, end=" ")
        
```
